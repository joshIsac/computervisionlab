```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from scipy.ndimage import convolve
import matplotlib.pyplot as plt
```

```python
src_img= cv2.imread('/content/drive/MyDrive/ferrari.jpg',cv2.IMREAD_GRAYSCALE)
cv2_imshow(src_img)
```

```python
img = cv2.imread('/content/drive/MyDrive/ferrari.jpg')
cv2_imshow(img)
```

```python
#identity kernel
identity_kernel = np.array([[0, 0, 0],[0, 1, 0], [0, 0, 0]])
img = cv2.filter2D(img, -1, identity_kernel)
```

The output image produced is exactly like the image that is given as the input. It does change the input image. It is a square matrix with the center element equal to 1. All the other elements of the matrix are 0.

```python
cv2_imshow(img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#edge detection
edge_detection_kernel = np.array([[-1, -1, -1],
                    [-1, 8, -1],
                    [-1, -1, -1]])

# Applying the filter2D() function
img = cv2.imread('/content/drive/MyDrive/ferrari.jpg')
img = cv2.filter2D(src=img, ddepth=-1, kernel=edge_detection_kernel)
cv2_imshow(src_img)
cv2_imshow(img)
```

```python
#sharpenning
shrp_kernel=np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
# Sharpeneded image is obtained using the variable sharp_img
# cv2.fliter2D() is the function used
# src is the source of image(here, img)
# ddepth is destination depth. -1 will mean output image will have same depth as input image
# kernel is used for specifying the kernel operation (here, sharp_kernel)
sharp_img = cv2.filter2D(src=img, ddepth=-1, kernel=shrp_kernel)
cv2_imshow(sharp_img)
```

```python
#boxblur kernel
box_blur_ker = np.array([[0.1111111, 0.1111111, 0.1111111],
                    [0.1111111, 0.1111111, 0.1111111],
                    [0.1111111, 0.1111111, 0.1111111]])
Box_blur = cv2.filter2D(src=img, ddepth=-1, kernel=box_blur_ker)
cv2_imshow(Box_blur)
```

This is obtained by uniformly averaging the values in the neighborhood. It is a straightforward blur. The values should be summed up to 1.

```python
#gaussian blur
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gaussian_blur = cv2.GaussianBlur(src=img, ksize=(3,3),sigmaX=0, sigmaY=0)
cv2_imshow(gaussian_blur)
```

This filter can be obtained by finding the weighted average of the pixel. The weights of the pixels are calculated on the basis of distance from the center of the kernel. Pixels nearer to the center of the kernel influence more on the weighted average.
