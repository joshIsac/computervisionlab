```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
src_img= cv2.imread("/content/drive/MyDrive/ferrari.jpg",cv2.IMREAD_GRAYSCALE)
cv2_imshow(src_img)
```

```python
kernel = np.array([
  [10, 25, 1],
  [10, 1, 10],
  [20, 14, 20]
])
img = cv2.filter2D(src_img, -1, kernel)
```

```python
cv2_imshow("original image", src_img)
cv2_imshow("filter2d image", img)
cv2.imwrite("Filter2d Blur Image.jpg", img)
cv2.waitKey()
cv2.destroyAllWindows()
```
