Affine transformation :is A transformation that can be expressed in the form of a matrix multiplication  (linear transformation) followed by a vector addition (translation).
we can use an Affine Transformation to express:

Rotations (linear transformation)
Translations (vector addition)
Scale operations (linear transformation)
you can see that, in essence, an Affine Transformation represents a relation between two images.
it is represented int the form of 2*3 matrix

```python
#Using ImageIO
import imageio as iio

# read an image
img = iio.imread('/content/drive/MyDrive/pagani huayra.jpg')

# write it in a new format
iio.imwrite("pagani Huayra.png", img)
```

```python
#converting the img to gray using rgb2bgr
img = rgb2gray(imread("pagani Huayra.png"))
w, h = img.shape
print("width:",w)
print("height:",h)
```

```python
#identity transform
ident=np.array([[1,0,0],[0,1,0],[0,0,1]])
img1 = ndi.affine_transform(img, ident)
#plotting the image
plt.imshow(img1)
```

```python
#Apply reflection transform :a photographic method that captures a subject's shape and colour to reveal surface information invisible under normal examination.
translate=np.array([[1,0,100],[0,1,150],[0,0,1]])
img2=ndi.affine_transform(img,translate)
#plotting the image
plt.imshow(img2)
```

```python
#visualize an image using the scikit-image (skimage)
from skimage import io
import matplotlib.pyplot as  plt
# Load the image
image = io.imread(image_path)
# Display the image
plt.imshow(image)
plt.show()
```

```python
#resizing the image
from skimage import transform,io,util
pag_image=io.imread("pagani Huayra.png")
resized_image = transform.resize(pag_image, (300, 500))
# Convert the image data to uint8 format
resized_image_ubyte = util.img_as_ubyte(resized_image)

# Save the resized image
io.imsave("resized_pagani_huayra.png", resized_image_ubyte)
```
